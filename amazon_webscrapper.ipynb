Fetching Reviews from Amazon for all the devices

```python
rate_links=[]
with open('ratings_links.txt') as csvfile:
    csvReader = csv.reader(csvfile)
    for row in csvReader:
        if row ==[]:
            break
        rate_links.append(row[0])
        print(row[0])

for url in rate_links:
    reviews_bank=[]
    ratings_bank=[]
    
    for j in range(1,101):
        url_n = url+str(j)
        print(url_n)
        req = Request(url_n , headers={'User-Agent': 'Mozilla/5.0'})
        webpage = urlopen(req).read()
        soup = BeautifulSoup(webpage,'html.parser')
        
        reviews=soup.find_all('span',{'data-hook':'review-body'})
        print(len(reviews))
        review_content = []
        for i in range(0,len(reviews)):
            review_content.append(reviews[i].get_text())
        review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
        review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
        
        reviews_bank.append(review_content)
        
        #review Rating
        rating = soup.find_all('i',class_='review-rating')

        rate = []
        for i in range(0,len(rating)):
            rate.append(rating[i].get_text())
        if rate:
            rate.pop(0)
            rate.pop(0)
        print(len(rate))
        ratings_bank.append(rate)

    reviews = np.concatenate(reviews_bank)
    ratings= np.concatenate(ratings_bank)

    
```
